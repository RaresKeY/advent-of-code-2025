# Day 3: Lobby - Part 1

```python
batteries = range(1, 10)

class Bank():
    def __init__(self, file_path: str):
        self.banks: list[str] = []
        self.pairs: list[int] = []

        self.total = 0

        self.read_banks(file_path)
        self.get_pairs()

    def read_banks(self, file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f.readlines():
                self.banks.append([int(x) for x in line.strip()])

    def get_pairs(self):
        for b in self.banks:
            n1 = max(b)
            index1 = b.index(n1)
            b[index1] = -1

            n2 = max(b)
            index2 = b.index(n2)
            b[index2] = -1

            final = int(str(n1) + str(n2)) if index1 <= index2 else int(str(n2) + str(n1))

            n3 = max(b[index1:])
            f2 = n1 * 10 + n3 if n3 >= 0 else 0

            self.total += final if final > f2 else f2

    def __str__(self):
        return str(self.banks)

bat_bank = Bank("batteries.csv")
print(bat_bank.total)
```

# Part 2 - monotonic decreasing stack

```python
class Bank():
    def __init__(self, file_path: str):
        self.banks: list[str] = []
        self.pairs: list[int] = []

        self.total = 0

        self._read_banks(file_path)
        self._monotonic_decreasing()

    def _read_banks(self, file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f.readlines():
                self.banks.append([int(x) for x in line.strip()])
        
    def _monotonic_decreasing(self, num=12):
        for b in self.banks:
            k = num
            stack = []

            for i, digit in enumerate(b):
                while stack and stack[-1][1] < digit and len(stack) + (len(b) - i) > k:
                    stack.pop()
                if len(stack) < k:
                    stack.append((i, digit))

            mask = [0] * len(b)
            for idx, _ in stack:
                mask[idx] = 1

            final = [str(x) for i, x in enumerate(b) if mask[i]]
            self.total += int(''.join(final))

bat_bank = Bank("batteries.csv")
print(bat_bank.total)
```
