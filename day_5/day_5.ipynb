# Day 5 - Part 1

```python
ing_ranges = []
data = []
to_list = ing_ranges

with open("ingredients.txt", 'r', encoding='utf-8') as f:
    for l in f.readlines():
        if l == '\n':
            to_list = data
            continue

        to_list.append(l.strip())

ing_ranges = [list(map(int, x.split('-'))) for x in ing_ranges]
ing_ranges.sort(key=lambda x: x[0])

x = 0
while x < len(ing_ranges) - 1:
    if ing_ranges[x][1] >= ing_ranges[x + 1][0]:
        ing_ranges[x][1] = max(ing_ranges[x][1], ing_ranges[x+1][1])
        ing_ranges.pop(x + 1)
    else:
        x += 1

data = [int(x) for x in data]

eff_ranges = [range(x, y + 1) for x, y in ing_ranges]
fresh = set((n for n in data for r in eff_ranges if n in r))

print(len(fresh))
```

# Day 5 - Part 2

```python
ing_ranges = []
data = []
to_list = ing_ranges

with open("ingredients.txt", 'r', encoding='utf-8') as f:
    for l in f.readlines():
        if l == '\n':
            to_list = data
            continue

        to_list.append(l.strip())

ing_ranges = [list(map(int, x.split('-'))) for x in ing_ranges]
ing_ranges.sort(key=lambda x: x[0])

x = 0
while x < len(ing_ranges) - 1:
    if ing_ranges[x][1] >= ing_ranges[x + 1][0]:
        ing_ranges[x][1] = max(ing_ranges[x][1], ing_ranges[x+1][1])
        ing_ranges.pop(x + 1)
    else:
        x += 1

fresh_ranges = [(y + 1) - x for x, y in ing_ranges]

print(sum(fresh_ranges))
```
