```python
class Beam:
    children = []
    count = 0
    splits = 0
    split_history = []
    def __init__(self, manifold, position):
        self.position = position
        self.manifold = manifold
        self.destroyed = False
        Beam.count += 1

        self.travel()

    def travel(self):
        while self.position[0] < len(self.manifold) - 1:
            row = self.position[0]
            col = self.position[1]
            line = self.manifold[row]
            self.manifold[row] = line[:col] + '|' + line[col + 1:]
            if self.manifold[row + 1][col] == '^':
                split1 = self.split((row + 1, col + 1))
                split2 = self.split((row + 1, col - 1))

                Beam.splits += 1 if split1 or split2 else 0

                self.position = (row + 1, col)
                self.destroyed = True
                break

            self.position = (row + 1, col)

    def split(self, splitter_pos):
        if splitter_pos in Beam.split_history:
            return False

        self.children.append(Beam(self.manifold, splitter_pos))
        Beam.split_history.append(splitter_pos)
        return True

with open("manifold.txt", 'r', encoding='utf-8') as f:
    manifold = [line.strip() for line in f.readlines()]

start_pos = (0, manifold[0].index('S'))
og_beam = Beam(manifold, start_pos)

for b in set([t.position for t in og_beam.children if not t.destroyed]):
    line = manifold[b[0]]
    manifold[b[0]] = line[:b[1]] + '|' + line[b[1]+1:]

# for line in manifold:
#     print(line)

mask = [1 if c == '|' else 0 for c in manifold[-1]]
print("beams:", Beam.count, "splits:", Beam.splits, "final_beam_count:", sum(mask))
```
