# Day 7 - Part 1

```python
class Beam:
    count = 0
    splits = 0
    split_history = []

    def __init__(self, manifold, position):
        self.position = position
        self.manifold = manifold

        Beam.count += 1

        self.travel()

    def travel(self):
        while self.position[0] < len(self.manifold) - 1:
            row = self.position[0]
            col = self.position[1]
            if self.manifold[row + 1][col] == '^':
                split1 = self.split((row + 1, col + 1))
                split2 = self.split((row + 1, col - 1))

                Beam.splits += 1 if split1 or split2 else 0

                self.position = (row + 1, col)
                break

            self.position = (row + 1, col)

    def split(self, splitter_pos):
        if splitter_pos in Beam.split_history:
            return False

        Beam(self.manifold, splitter_pos)
        Beam.split_history.append(splitter_pos)
        return True

with open("manifold.txt", 'r', encoding='utf-8') as f:
    manifold = [line.strip() for line in f.readlines()]

start_pos = (0, manifold[0].index('S'))
og_beam = Beam(manifold, start_pos)

print("beams:", Beam.count, "splits:", Beam.splits)
```

# Day 7 - Part 2

```python
with open("manifold.txt", 'r', encoding='utf-8') as f:
    manifold = [line.strip() for line in f.readlines()]

manifold[0] = manifold[0].replace('S', '1')

for li, line in enumerate(manifold):
    manifold[li] = line.replace('.', '0')
    # print(manifold[li])

manifold = [[c for c in x] for x in manifold]

for li in range(1, len(manifold)):
    for ci in range(len(manifold[0])):
        og_dimensions = manifold[li - 1][ci]
        selection = manifold[li][ci]

        if selection == '^':
            manifold[li][ci - 1] = str(int(manifold[li][ci - 1]) + int(og_dimensions))
            manifold[li][ci + 1] = str(int(manifold[li][ci + 1]) + int(og_dimensions))

        elif selection.isdigit() and og_dimensions.isdigit():
            manifold[li][ci] = str(int(selection) + int(og_dimensions))

print(sum([int(x) for x in manifold[-1]]))
```

# This takes 61,600 years to compute

```python
class Beam:
    count = 0
    branch_stack = []

    def __init__(self, manifold, position):
        self.position = position
        self.manifold = manifold

        Beam.count += 1
        self.travel()

    def travel(self):
        while self.position[0] < len(self.manifold) - 1:
            row = self.position[0]
            col = self.position[1]

            if self.manifold[row + 1][col] == '^':
                self.position = (row + 1, col - 1)
                Beam.branch_stack.append((row + 1, col + 1))
            else:
                self.position = (row + 1, col)

with open("manifold.txt", 'r', encoding='utf-8') as f:
    manifold = [line.strip() for line in f.readlines()]

start_pos = (0, manifold[0].index('S'))
og_beam = Beam(manifold, start_pos)

while Beam.branch_stack:
    Beam(manifold, Beam.branch_stack[-1])

print("beams/timelines:", Beam.count)
```
