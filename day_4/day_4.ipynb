# Day 1 - Part 1

```python
data = []

with open("paper_rolls.txt", 'r', encoding='utf-8') as f:
    for line in f.readlines():
        data.append(line.strip())

def convert(data):
    data = [[x for x in line] for line in data]
    new_data = [['0' if x == '@' else 9 for x in line] for line in data]

    len_data = len(data)
    len_line = len(data[0])

    for li, line in enumerate(data):
        for ci, c in enumerate(line):
            if c == '@':
                ls = max(li - 1, 0)
                le = min(li + 1, len_data - 1)

                cs = max(ci - 1, 0)
                ce = min(ci + 1, len_line - 1)

                count = 0
                for i1 in range(ls, le + 1):
                    for i2 in range(cs, ce + 1):
                        if data[i1][i2] == '@':
                            count += 1

                new_data[li][ci] = count - 1

    return new_data

total = 0
remove_count = 1


new_data = convert(data)
hits = [[1 if x < 4 else 0 for x in line] for line in new_data]
data = [['x' if hits[di][li] == 1 else x for li, x in enumerate(line)] for di, line in enumerate(data)]
remove_count = sum([sum(x) for x in hits])
total += remove_count

print(total)
```
