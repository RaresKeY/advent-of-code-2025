# Day 6 - Part 1

```python
math_sheet = []

with open("math_sheet.txt", 'r', encoding='utf-8') as f:
    for line in f.readlines():
        math_sheet.append([x.strip() for x in line.split(' ') if x and x != '\n'])
        
problems = []

for i in range(len(math_sheet[0])):
    data = []
    for n in range(len(math_sheet) - 1):
        data.append(math_sheet[n][i])
    problems.append({"numbers": data, "sign": math_sheet[-1][i]})

total = 0

for l in problems:
    arranged = l["sign"].join(l["numbers"])
    total += eval(arranged)

print(total)
```

# Day 6 - Part 2

```python
math_sheet = []

with open("math_sheet.txt", 'r', encoding='utf-8') as f:
    for line in f.readlines():
        math_sheet.append(list(line.strip('\n')))

# pad
for y in range(len(math_sheet[0])):
    sep = False
    for i in range(len(math_sheet) - 1):
        if math_sheet[i][y] != ' ':
            sep = True
    if sep:
        for i in range(len(math_sheet) - 1):
            if math_sheet[i][y] == ' ':
                math_sheet[i][y] = 'x'

math_sheet = [''.join(x).split() for x in math_sheet]

problems = []

for i in range(len(math_sheet[0])):
    data = []
    for n in range(len(math_sheet) - 1):
        data.append(math_sheet[n][i])
    problems.append({"numbers": data, "sign": math_sheet[-1][i]})

for p in problems:
    new_column = []
    for i in range(len(p["numbers"][0])):
        new_x = ''
        for n in p["numbers"]:
            new_x += n[i] if n[i] != 'x' else ''
        new_column.append(new_x)

    p["numbers"] = new_column[::-1]

total = 0

for l in problems:
    arranged = l["sign"].join(l["numbers"])
    total += eval(arranged)

print(total)
```
