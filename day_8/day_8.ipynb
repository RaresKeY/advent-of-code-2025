# Day 8 - Part 1

```python
import math

class Grouping():
    children = []
    def __init__(self, a, b):
        self.add_circuit(a, b)

    def add_circuit(self, a, b):
        ai, bi = -1, -1
        for i, c in enumerate(Grouping.children):
            ai = i if a in c else ai
            bi = i if b in c else bi

        if ai == bi:
            return

        a = {'circuit': Grouping.children[ai], 'len': len(Grouping.children[ai]), 'index': ai}
        b = {'circuit': Grouping.children[bi], 'len': len(Grouping.children[bi]), 'index': bi}

        Grouping.children[min(a['index'], b['index'])].extend(Grouping.children.pop(max(a['index'], b['index'])))


with open("junctions.txt", 'r', encoding='utf-8') as f:
    junctions = [tuple(map(int, line.strip().split(','))) for line in f.readlines()]

def distance_3d(a, b):
    x1, y1, z1 = a
    x2, y2, z2 = b

    dx, dy, dz = abs(x1 - x2), abs(y1 - y2), abs(z1 - z2)
    d = (dx ** 2 + dy ** 2 + dz ** 2) ** 0.5

    return d

d_pairs = dict()
for moving_group in range(len(junctions)):
    for y in range(moving_group + 1, len(junctions)):
        d = distance_3d(junctions[moving_group], junctions[y])
        d_pairs[(moving_group, y)] = d

for j in junctions:
    Grouping.children.append([j])

sorted_dict = dict(sorted(d_pairs.items(), key=lambda x: x[1]))
sliced_dict = dict(list(sorted_dict.items())[:1000])

for k, v in sliced_dict.items():
    a, b = junctions[k[0]], junctions[k[1]]
    # print(a, b)
    Grouping(a, b)


l = [len(s) for s in Grouping.children]
l.sort(reverse=True)
print(math.prod(l[:3]))
```

# Day 8 - Part 2

```python
from dataclasses import dataclass

@dataclass
class Circuit:
    group: int
    coord: tuple

class Grouping():
    def __init__(self, junctions):
        n = len(junctions)
        self.circuits = [Circuit(x, junctions[x]) for x in range(n)]
        self.gn = len(junctions)

    def join(self, ai, bi):
        target_group = self.circuits[ai].group
        moving_group = self.circuits[bi].group

        if target_group == moving_group:
            return

        for c in self.circuits:
            if c.group == moving_group:
                c.group = target_group
        
        self.gn -= 1


with open("junctions.txt", 'r', encoding='utf-8') as f:
    junctions = [tuple(map(int, line.strip().split(','))) for line in f.readlines()]

circuits = Grouping(junctions)

def distance_3d(a, b):
    x1, y1, z1 = a
    x2, y2, z2 = b

    dx, dy, dz = x1 - x2, y1 - y2, z1 - z2
    d = (dx ** 2 + dy ** 2 + dz ** 2) ** 0.5

    return d

d_pairs = []
for i in range(len(junctions)):
    for y in range(i + 1, len(junctions)):
        d = distance_3d(junctions[i], junctions[y])
        d_pairs.append((d, i, y))

d_pairs.sort(key=lambda x: x[0])

count = 0
for d, a, b in d_pairs:
    count += 1
    circuits.join(a, b)

    if circuits.gn == 1:
        x1 = circuits.circuits[a].coord[0]
        x2 = circuits.circuits[b].coord[0]
        print("ANSWER:", x1 * x2, "steps:", count)
        break
```
